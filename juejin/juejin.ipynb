对前100的文章进行分词，分组统计处理，
看看那些关键字是大家更愿意看的

```python
import pandas as pd
df = pd.read_csv("./entities.csv")
            
df = df.sort_values(by=["viewsCount","commentsCount"], ascending=False)

data = []
top_df = df[:100]
for index,row in top_df.iterrows():
    title = (row['title'])
    import jieba.analyse
    jieba.analyse.set_stop_words("./stopwords.txt")
#     keywords = jieba.cut(title, cut_all=False)
    keywords = jieba.analyse.extract_tags(title, topK=10)
    for kw in keywords:
        if kw not in ["Android", "前端","Java", "Google",'完整']:
            data.append([kw, 1])

from pandas import DataFrame
import numpy as np
np.array(data)
keyword_df = DataFrame(data, columns=["keyword", "viewsCount"])
# print(keyword_df)
keyword_df = keyword_df.groupby(['keyword']).sum().sort_values(by="viewsCount", ascending=False)
# groupby_regiment = keyword_df['viewsCount'].groupby(keyword_df['keyword'])
# groupby_regiment.mean()
keyword_df = keyword_df[:20]
keyword_df.index.values
import plotly.plotly as py
import plotly.graph_objs as go


trace3 = go.Bar(
    x=keyword_df.index.values,
    y=keyword_df['viewsCount'],
    name='访问'
)

data = [trace3]
layout = go.Layout(
    barmode='group'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')


    
# top_df
```
